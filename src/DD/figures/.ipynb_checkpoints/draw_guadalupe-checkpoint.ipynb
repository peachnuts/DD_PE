```python
qft_circuit_names = ['QFT4', 'QFT5', 'QFT6', 'QFT7', 'QFT8', 'QFT9', 'QFT10', 'QFT11', 'QFT12']
qft_results_data_avg = {}
qft_results_data = [qft_results_data1, qft_results_data2, qft_results_data3]
num = 11 #number of PSTs for one job

for circuit_name in qft_circuit_names:
    psts = []
    qft_results_data_avg[circuit_name] = []
    for i in range(num):
        avg = sum([qft_res[circuit_name][i] for qft_res in qft_results_data]) / 3
        qft_results_data_avg[circuit_name].append(avg)
```

```python
import matplotlib.pyplot as plt
baseline = []

qft_data = []
for name in qft_circuit_names:
    result_data = qft_results_data_avg[name]
    baseline.append(result_data[0])
    ratio = [result_data[0]/i for i in result_data[1:]]
    qft_data.append(ratio)


qft_data = np.array(qft_data)

# hahn_X = data[:, 0]
# hahn_Y = data[:, 1]
cp = qft_data[:, 2]
cpmg = qft_data[:, 3]
xy4 = qft_data[:, 4]
xy8 = qft_data[:, 5]
xy16 = qft_data[:, 6]
udd_X = qft_data[:, 7]
udd_Y = qft_data[:, 8]
kdd = qft_data[:, 9]

x_ticks = qft_circuit_names
X = np.arange(len(x_ticks))

from matplotlib.pyplot import figure
fig = figure(num=None, figsize=(8, 6), dpi=600, facecolor='w', edgecolor='k')
ax = fig.add_axes([0,0,1,1])

# plt.plot(X, hahn_Y, linestyle='--', marker='o', color='g', label='hahn_X')
# plt.plot(X, hahn_X, linestyle='--', marker='s', color='m', label='hahn_Y')
plt.plot(X, cp, linestyle='--', marker='o', color='orange', label='cp')
plt.plot(X, cpmg, linestyle='--', marker='.', color='b', label='cpmg')
plt.plot(X, xy4, linestyle='--', marker='^', color='r', label='xy4')
plt.plot(X, xy8, linestyle='--', marker='X', color='g', label='xy8')
plt.plot(X, xy16, linestyle='--', marker='h', color='c', label='xy16')
plt.plot(X, udd_X, linestyle='--', marker='d', color='y', label='udd_x')
plt.plot(X, udd_Y, linestyle='--', marker='+', color='k', label='udd_y')
plt.plot(X, kdd, linestyle='--', marker='*', color='m', label='kdd')
plt.axhline(1, color='tab:brown', lw=2)
plt.legend(loc='best', fontsize=16)
# ax.set_title('Relative JSD result for QFT circuit on ibmq_guadalupe', fontsize=18)
ax.set_xticks(X)
ax.set_xticklabels(x_ticks, fontsize=14)
ax.set_ylabel('Relative JSD', fontsize=16)

plt.savefig('TQE_Relat_JSD_QFT_guadalupe_avg.pdf', bbox_inches='tight', pad_inches=0)
```
